**Missing Data**

```python
missing=df.isna().sum().sort_values(ascending=False)
missing.plot.bar(figsize=(10,6))
plt.xlabel('Columns with missing values')
plt.ylabel('Count')
plt.show()
```

```python
missing
```

```python
#filling missing data
df=df.fillna(method='ffill')
df
```
